## Imóveis Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
```

```python
dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
residencial
```

```python
# Vamos checar se a Series bate com os dados anteriores:
dados['Tipo'].isin(residencial).head(10)
```

```python
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
# O processo aqui foi criar um novo DataFrame com os valores booleanos de um Series, assim agregando
# ao novo DataFrame somente os valores verdadeiros
dados_residencial = dados[selecao]
dados_residencial
```

```python
# Checando o novo DataFrame com o drop_duplicates:
dados_residencial['Tipo'].drop_duplicates()
```

```python
# A diferença do número de linhas antes e depois
dados_residencial.shape
```

```python
dados.shape
```

```python
# Recriando os index com range() já que mudamos o DataFrame
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```

## Extra
### Organizando dataframes

```python
data = [[1,2,3], [4,5,6], [7,8,9]]
data
```

```python
# Cada caractere um item da lista
list('321')
```

```python
# Criando DataFrame com método list
df = pd.DataFrame(data, list('321'), list('ZYX'))
df
```

```python
# Reorganizar ps index do DataFrame
df.sort_index(inplace = True)
df
```

```python
# Vamos fazer isso também para as colunas
df.sort_index(inplace = True, axis = 1)
df
```

```python
df.sort_values(by = 'X', inplace= True)
# entendi porra nenhuma
```

```python
df.sort_index()
df.sort_index(axis = 1)
df
```
