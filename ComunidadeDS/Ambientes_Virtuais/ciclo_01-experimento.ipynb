***if* and *elif* statements**

```python
# Define response code
response_code = 400
```

```python
# print report according to error code
if response_code == 100:
    print("Informational error!")
elif response_code == 200:
    print("Success!")
elif response_code == 300:
    print("Redirected!")
elif response_code == 400:
    print("Client error!")
elif response_code == 500:
    print("Server error!")
```

***match* and *case* statements**

*only available on python > 3.10*

```python
# print report according to error code
match response_code:
    case 100:
        print("Informational error!")
    case 200:
        print("Success!")
    case 300:
        print("Redirected!")
    case 400:
        print("Client error!")
    case 500:
        print("Server error!")
```
