```python
import pandas as pd
import re
df_root = pd.read_csv('C:/Users/lui-m/Documents/GitHub/DS_Notebooks/ComunidadeDS/FTC_Python/datasets/train.csv')
df_root.head()
```

```python
# Fazendo cópia do dataframe lido
df = df_root.copy()

# Removendo espaço da string
for i in range(len(df)):
    df.loc[i, 'ID'] = df.loc[i, 'ID'].strip()
    df.loc[i, 'Delivery_person_ID'] = df.loc[i, 'Delivery_person_ID'].strip()
    df.loc[i, 'Road_traffic_density'] = df.loc[i, 'Road_traffic_density'].strip()
    df.loc[i, 'Type_of_order'] = df.loc[i, 'Type_of_order'].strip()
    df.loc[i, 'Type_of_vehicle'] = df.loc[i, 'Type_of_vehicle'].strip()
    
# Excluindo as linhas com a idade dos entregadores vazia
linhas_nao_vazias = df['Delivery_person_Age'] != 'NaN '
df = df.loc[linhas_nao_vazias, :]

# Conversao de texto/categoria/string para numeros inteiros
df['Delivery_person_Age'] = df['Delivery_person_Age'].astype(int)

# Conversao de texto/categoria/string para numeros decimais
df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].astype(float)

# Conversao de texto para data
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')

# Remove as linhas da coluna multiple_deliveries que tenham o conteúdo igual a 'NqN ' e fazendo o cast para inteiro
linhas_nao_vazias = df['multiple_deliveries'] != 'NaN '
df = df.loc[linhas_nao_vazias, :]
df['multiple_deliveries'] = df['multiple_deliveries'].astype(int)

# Removendo texto de números e resetando index
df = df.reset_index(drop=True)
for i in range(len(df)):
    df.loc[i, 'Time_taken(min)'] = re.findall(r'\d+', df.loc[i, 'Time_taken(min)'])
```

### 15. Qual o tipo de veículo que mais fez entregas com a condição de trânsito pesado? ( High )

```python
tipos_veiculos = list(df['Type_of_vehicle'].unique())
total_entregas_15 = 0
tipo_veiculo_15 = ''

for veiculo in tipos_veiculos:
    filtro = (df['Type_of_vehicle'] == veiculo) & (df['Road_traffic_density'] == 'High')
    
    tipo_veiculo_entregas = df.loc[filtro, :].shape[0]
    
    print(f'{veiculo} -> {tipo_veiculo_entregas}')
    
    if total_entregas_15 < tipo_veiculo_entregas:
        total_entregas_15 = tipo_veiculo_entregas
        tipo_veiculo_15 = veiculo
        
print(f'Maior : {tipo_veiculo_15} -> {total_entregas_15}')
```

### 16. Qual o tipo de pedido mais feito durante condições climáticas de tempestade de areia?


```python
tipos_pedidos = list(df['Type_of_order'].unique())
total_entregas = 0
tipo_pedido = ''

for pedido in tipos_pedidos:
    filtro = (df['Type_of_order'] == pedido) & (df['Weatherconditions'] == 'conditions Sandstorms')
    
    tipo_pedido_entregas = df.loc[filtro, :].shape[0]
    
    print(f'{pedido} -> {tipo_pedido_entregas}')
    if total_entregas < tipo_pedido_entregas:
        total_entregas = tipo_pedido_entregas
        tipo_pedido = pedido
        
print(f'O tipo de pedido mais feito durante tempestade de areia é o {tipo_pedido}, com um total de {total_entregas} entregas')
```

### 17. Qual a cidade com o maior número de pedidos de Bebidas (Drinks) feito em Scooter?

```python
cidades = list(df['City'].unique())
mais_pedidos = 0
maior_cidade = ''
for cidade in cidades:
    print(cidade)
    filtro = (df['City'] == cidade) & (df['Type_of_vehicle'] == 'scooter') & (df['Type_of_order'] == 'Drinks')
    
    cidade_drinks_scooter = df.loc[filtro, :].shape[0]
    if cidade_drinks_scooter > mais_pedidos:
        mais_pedidos = cidade_drinks_scooter
        maior_cidade = cidade
print(f'A cidade com maior número de pedidos Drinks feitos em Scooter foi: {maior_cidade} com {mais_pedidos} pedidos')
```
