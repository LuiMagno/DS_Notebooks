### 3. Qual a idade do entregador mais velho? E do mais novo?

```python
# Vamos limpar os NaN
linhas_limpas = df['Delivery_person_Age'] != 'NaN '
df1 = df.loc[linhas_limpas, :]

idade_maior = df1['Delivery_person_Age'].max()
idade_menor = df1['Delivery_person_Age'].min()

print(f'A idade do entregador mais velho é {idade_maior}')
print(f'A idade do entregador mais novo é {idade_menor}')
```

### 4. Qual o ID do entregador com a maior idade? E o ID do entregaor com a menor idade?

```python
linhas_maiores = df1['Delivery_person_Age'] == idade_maior
df2 = df1.loc[linhas_maiores, :]
df2[['Delivery_person_ID', 'Delivery_person_Age']]

linhas_menores = df1['Delivery_person_Age'] == idade_menor
df3 = df1.loc[linhas_menores, :]
df3[['Delivery_person_ID', 'Delivery_person_Age']]
```
