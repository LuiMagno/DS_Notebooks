```python
import pandas as pd
import re

df_root = pd.read_csv('C:/Users/lui-m/Documents/GitHub/DS_Notebooks/ComunidadeDS/FTC_Python/datasets/train.csv')
df_root.head()
```

```python
# Fazendo cópia do dataframe lido
df = df_root.copy()

# Removendo espaço da string
for i in range(len(df)):
    df.loc[i, 'ID'] = df.loc[i, 'ID'].strip()
    df.loc[i, 'Delivery_person_ID'] = df.loc[i, 'Delivery_person_ID'].strip()
    
# Excluindo as linhas com a idade dos entregadores vazia
linhas_nao_vazias = df['Delivery_person_Age'] != 'NaN '
df = df.loc[linhas_nao_vazias, :]

# Conversao de texto/categoria/string para numeros inteiros
df['Delivery_person_Age'] = df['Delivery_person_Age'].astype(int)

# Conversao de texto/categoria/string para numeros decimais
df['Delivery_person_Ratings'] = df['Delivery_person_Ratings'].astype(float)

# Conversao de texto para data
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')

# Remove as linhas da coluna multiple_deliveries que tenham o conteúdo igual a 'NqN ' e fazendo o cast para inteiro
linhas_nao_vazias = df['multiple_deliveries'] != 'NaN '
df = df.loc[linhas_nao_vazias, :]
df['multiple_deliveries'] = df['multiple_deliveries'].astype(int)

# Removendo texto de números e resetando index
df = df.reset_index(drop=True)
for i in range(len(df)):
    df.loc[i, 'Time_taken(min)'] = re.findall(r'\d+', df.loc[i, 'Time_taken(min)'])
```

```python
# 9. Quantos tipos de densidade de trânsito existem na base de dados? Quais os nomes delas?
condicoes_climaticas = list(df['Weatherconditions'].unique())

for condicao in condicoes_climaticas:
    numero_entregas = df.loc[df['Weatherconditions'] == condicao, :].shape[0]
    print(f'O numero de entregas feitas na densidade de transito {condicao} foi de {numero_entregas}')
```

```python
# 10. Quantos entregadores únicos fizeram entregas em cada condição climática?
for condicao in condicoes_climaticas:
    entregadores_unicos_clima = df.loc[df['Weatherconditions'] == condicao, 'Delivery_person_ID'].nunique()
    print(f'{entregadores_unicos_clima} fizeram entregas na condicao climática {condicao}')
```

```python
#11. Quantas entregas foram feitas em cada tipo de densidade de trânsito?
densidades_transito = list(df['Road_traffic_density'].unique())
for densidade in densidades_transito:
    entregas_feitas = len(df.loc[df['Road_traffic_density'] == densidade])
    print(f'{entregas_feitas} foram feitas na densidade de trânsito {densidade}')
```

```python
#12. Quantos entregadores únicos fizeram entregas em cada tipo de densidade de trânsito?
for densidade in densidades_transito:
    entregadores_unicos = df.loc[df['Road_traffic_density'] == densidade, 'Delivery_person_ID'].nunique()
    print(f'{entregadores_unicos} entregadores únicos fizeram entregas no tipo de densidade {densidade}')
```

```python
# 15. Qual o tipo de veículo que mais vez entregas com a condição de trânsito pesado? ( High )
tipos_veiculoss = list(df['Type_of_vehicle'].unique())

total_entregas = 0
tipo_veiculo = ''

for veiculo in tipos_veiculoss:
    filtro = (df['Type_of_vehicle'] == veiculo) & (df['Road_traffic_density'] == 'High')
    
    tipo_veiculo_entregas = df.loc[filtro, :].shape[0]
    
    print(f'{veiculo} -> {tipo_veiculo_entregas}')
    if total_entregas < tipo_veiculo_entregas:
        total_entregas = tipo_veiculo_entregas
        tipo_veiculo = veiculo
        
print(f'{tipo_veiculo} e {total_entregas}')
```
